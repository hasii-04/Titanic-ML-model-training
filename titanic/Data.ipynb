```python
import pandas as pd
import numpy as np

titanic_train = pd.read_csv('train.csv')
titanic_test = pd.read_csv('test.csv')

```

```python
missing_values_count_train = titanic_train.isnull().sum()
missing_values_count_test = titanic_test.isnull().sum()
missing_values_count_train[0:12]
missing_values_count_test[0:12]
```

```python
titanic_train.drop(['Cabin', 'Ticket', 'Name'], axis=1, inplace=True)
titanic_test.drop(['Cabin', 'Ticket', 'Name'], axis=1, inplace=True)
```

```python
titanic_train.dropna()
titanic_test.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
titanic_train['Sex'] = le.fit_transform(titanic_train['Sex'])
titanic_test['Sex'] = le.fit_transform(titanic_test['Sex'])
```

```python
titanic_train = pd.get_dummies(titanic_train, columns=['Embarked'], drop_first=True)
titanic_test = pd.get_dummies(titanic_test, columns=['Embarked'], drop_first=True)
titanic_train.head()
```
